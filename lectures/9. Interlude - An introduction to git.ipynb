Interlude - An introduction to git
==================================

We have already learned about Python, Numpy, Scipy, and Matplotlib, so you can already write code to carry out scientific analysis. When doing research, we just use combinations of these packages (and sometimes domain-specific packages, for example astronomy-specific packages) to analyze data and obtain scientific results. As your analysis gets more complicated, so will your code, and it is inevitable that you will introduce *bugs* into the code. A *bug* is essentially a mistake in the code that is causing results to be incorrect.

One very important concept in programming is called *version control* - this basically means that we record changes to the program over time. That way, if it stops working, then we can always go back to a previous version that we know worked. You can think of it as keeping a backup of the code every time you make a significant change.

There are several available tools to do version control, but in this lecture I will show you a little about a tool called [git](http://git-scm.com/). The basic syntax is:

    $ git <command> <options>

where ``$`` indicates we are typing the commands in the normal command-line, *not* IPython.

Setting up git
--------------

Before you start using ``git`` you need to use the following two commands:

    $ git config --global user.name "Your Name Here"
    $ git config --global user.email "your_email@example.com"

to tell ``git`` who you are.

Basics of git
-------------

The first concept to know about is that of a *repository*, which you can think of as a special folder inside which we will be keeping the history of the code. You can start off by creating a normal directory, then going inside it:

    $ mkdir test_repo
    $ cd test_repo

Now initialize the repository:

    $ git init
    Initialized empty Git repository in /Users/tom/test_repo/.git/

The ``test_repo`` directory is now a repository. The only difference to a normal directory is that it includes a hidden folder called ``.git`` that keeps track of the history.

Now we have to create a file that we want to track the history for. You can create a Python script any way you usually do - if you want to do it all in the Terminal, you can type:

    $ pico test.py

then write a couple of lines of code, for example:

    # test program
    print "Hello, World!"

press control-X to exit and say yes to the question about saving the file. Once the file has been saved, you can type ``ls`` to make sure the file is there:

    $ ls
    test.py

Type ``git status`` to find out the current state of the repository:

    $ git status
    # On branch master
    #
    # Initial commit
    #
    # Untracked files:
    #   (use "git add <file>..." to include in what will be committed)
    #
    #	test.py
    nothing added to commit but untracked files present (use "git add" to track)

You will notice that ``test.py`` is an *untracked* file which means the history will *not* be saved. By default, not all files are tracked, only the ones added with:

    $ git add <file>

Let's add ``test.py`` to be tracked:

    $ git add test.py

If you now type

    $ git status

you will notice that the file is now in the category 'changes to be committed':

    $ git status
    # On branch master
    #
    # Initial commit
    #
    # Changes to be committed:
    #   (use "git rm --cached <file>..." to unstage)
    #
    #	new file:   test.py
    #

This means that it will be included in the next version. We can now finally *commit*, which means that we create a new version, or *revision*. To do this we use the ``commit`` command which is used as:

    $ git commit -m "<commit message here>"

In our case, we do:

    $ git commit -m "Added a new file"
    [master (root-commit) 71b5722] Added a new file
     1 files changed, 2 insertions(+), 0 deletions(-)
     create mode 100644 test.py

We can now look at the version history using:

    $ git log

and we will see the change we just made:

    $ git log
    commit 71b5722398bdf6246bcb8dd5b7b44ed00345a741
    Author: Thomas Robitaille <thomas.robitaille@gmail.com>
    Date:   Sun Jun 9 19:16:19 2013 +0200

        Added a new file

Now let's say that we want to modify the script and improve it. Let's change the ``test.py`` script to:

    # test program
    print "Hello, World!"
    print "This is now a much better script"

We can now do ``git status`` to see the current status of the repository:

    $ git status
    # On branch master
    # Changed but not updated:
    #   (use "git add <file>..." to update what will be committed)
    #   (use "git checkout -- <file>..." to discard changes in working directory)
    #
    #	modified:   test.py
    #
    no changes added to commit (use "git add" and/or "git commit -a")

You will see that it knows that ``test.py`` has been modified. We can view the changes with:

    $ git diff <file>

so in our case, to view the changes in ``test.py``, we do:

    $ git diff test.py
    diff --git a/test.py b/test.py
    index dbf4df1..f3039b9 100644
    --- a/test.py
    +++ b/test.py
    @@ -1,2 +1,3 @@
     # test program
     print "Hello, World!"
    +print "This is now a much better script"



You can see the ``+`` indicates new lines, and there is one new one here. We can now tell ``git`` that we want to include this change in the next revision (it will only includes changes you explicitly add):

    $ git add test.py
    $ git commit -m "Improved the script"
    [master 4bad055] Improved the script
     1 files changed, 1 insertions(+), 0 deletions(-)

If we check the history, we now see:

    $ git log
    commit 4bad05595afe9a4ce3609999d5ac4715a95a6d90
    Author: Thomas Robitaille <thomas.robitaille@gmail.com>
    Date:   Sun Jun 9 19:22:40 2013 +0200

        Improved the script

    commit 71b5722398bdf6246bcb8dd5b7b44ed00345a741
    Author: Thomas Robitaille <thomas.robitaille@gmail.com>
    Date:   Sun Jun 9 19:16:19 2013 +0200

        Added a new file

Publishing to a remote hosting service
--------------------------------------

When collaborating on a project, it can be useful to have the repository be available on a remote server. One such service is GitHub, and you should already have an account there since you have been using [gist.github.com](http://gist.github.com) to submit solutions! Go to [github.com](https://github.com/) and click on 'New repository' on the right. Give it a name (and a description if you like) then click on 'Create repository'. You will then be shown a page that looks something like this:

![GitHub](github.png)

Click on 'HTTP' in the middle, and copy the address that appears - in the above case, it is ``https://github.com/astrofrog/test_repo.git``. You can now type the following in your repository:

    $ git remote add origin https://github.com/astrofrog/test_repo.git

This means 'add a remote called ``origin`` and set the URL to ``https://github.com/astrofrog/test_repo.git``. The syntax is:

    $ git remote add <name> <url>

Then, do:

    $ git push origin master

which will push the latest version of your repository online. The ``master`` part refers to the name of the *branch* you are working on, which is the default *branch* (we won't go into more detail here though). Now go back to the website and refresh, and you will see your code there!

Getting a remote repository
---------------------------

In the same way that you can publish a repository online, you can also download one to your machine to work on. This is called *cloning*. For example, let's clone a repository that I've worked on in the past:

    $ git clone https://github.com/astrofrog/pyavm.git
    Cloning into pyavm...
    remote: Counting objects: 695, done.
    remote: Compressing objects: 100% (309/309), done.
    remote: Total 695 (delta 391), reused 675 (delta 373)
    Receiving objects: 100% (695/695), 556.64 KiB | 443 KiB/s, done.
    Resolving deltas: 100% (391/391), done.

We can now go into this directory and see the latest changes:

    $ cd pyavm
    $ ls
    CHANGES     LICENSE     MANIFEST.in README.md   pyavm       setup.py
    $ git log
    commit 550d0da923ba1f2d6b315d0ad57493f7395faee8
    Author: Thomas Robitaille <thomas.robitaille@gmail.com>
    Date:   Sun Apr 21 13:22:02 2013 +0200

        Updated version strings to 0.9.1

    commit 507ed0a6c980476370bbced411c3672118716bb9
    Merge: 567328e 0245704
    Author: Thomas Robitaille <thomas.robitaille@gmail.com>
    Date:   Sun Apr 21 03:43:47 2013 -0700

        Merge pull request #17 from astrofrog/brute-force-extraction
    
        Allow AVM to be extracted from any file using brute-force

    commit 024570477742f51b142d3848771c0e7feea7e44f
    Author: Thomas Robitaille <thomas.robitaille@gmail.com>
    Date:   Sun Apr 21 12:33:59 2013 +0200

        Support parsing from any arbitrary files - this uses brute-force scanning
    ...

If you want, you can make a change and then commit it. Note that the repository on your computer is separate from the online one, so the change will only be in your local version. To make the change available online too, you would need to do:

    $ git push origin master

but it will not work here, because you do not have permission to edit the online repository (only I can). Note that you don't need to add the remote here because it is automatically done when you ``clone`` a repository. You can view the current ``remotes`` with:

    $ git remote -v
    origin	https://github.com/astrofrog/pyavm.git (fetch)
    origin	https://github.com/astrofrog/pyavm.git (push)
